```python
data_df = pd.read_csv('/home/dai/Desktop/AI_220940128008/Deep Learning/Inputs/fifa_2019.csv')
data_df.shape
```

```python
# removing rows with position = null
data_df = data_df[data_df["Position"].notnull()]
data_df.head()
```

```python
# Following columns appear to be relevant for our analysis
rel_cols = ["Position", 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
            'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
            'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
            'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
            'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
            'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
            'GKKicking', 'GKPositioning', 'GKReflexes']
```

```python
goalkeeper = 'GK'
forward = ['ST', 'LW', 'RW', 'LF', 'RF', 'RS','LS', 'CF']
midfielder = ['CM','RCM','LCM', 'CDM','RDM','LDM', 'CAM', 'LAM', 'RAM', 'RM', 'LM']
defender = ['CB', 'RCB', 'LCB', 'LWB', 'RWB', 'LB', 'RB']
```

```python
#Assign labels to goalkeepers
data_df.loc[data_df["Position"] == "GK", "Position"] = 0

#Defenders
data_df.loc[data_df["Position"].isin(defender), "Position"] = 1

#Midfielders
data_df.loc[data_df["Position"].isin(midfielder), "Position"] = 2

#Forward
data_df.loc[data_df["Position"].isin(forward), "Position"] = 3

data_df['Position']= pd.to_numeric(data_df['Position'], downcast='integer')
```

```python
data_df = data_df[rel_cols]
data_df.head()
```

```python
y = data_df["Position"].values

X = data_df.drop("Position", axis = 1)

# one hot encoding
y = pd.get_dummies(y).values
X.shape, y.shape
```

```python
#  Split the data in training and test sets to measure performance of the model.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE )

print (X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print ('Is CUDA available: ', torch.cuda.is_available())

print ('CUDA version: ', torch.version.cuda )

print ('Current Device ID: ', torch.cuda.current_device())

print ('Name of the CUDA device: ', torch.cuda.get_device_name(torch.cuda.current_device()))
```

```python
# Get cpu or gpu device for training.

device = "cuda" if torch.cuda.is_available() else "cpu"

print(f"Using {device} device")
```

```python
class Model(nn.Module):
  
  def __init__(self, input_dim):

      ##Defining Layers
    super(Model, self).__init__()
    self.layer1=nn.Linear(input_dim, 16)
    self.layer2=nn.Linear(16, 4)
  
  def forward(self, x):
    x= torch.tanh(self.layer1(x))
    x= F.softmax(self.layer2(x), dim=1)
    return x

net= Model(X_train.shape[1]).to(device) ##Instantiate and move over to CUDA
print (net)
```

```python
# wrap up with Variable in pytorch and transfer them to GPU
train_X= Variable(torch.Tensor(X_train).float().to(device))
train_y= Variable(torch.Tensor(y_train).float().to(device))

test_X= Variable(torch.Tensor(X_test).float().to(device))
test_y= Variable(torch.Tensor(y_test).float().to(device))
```

```python
loss_fn= nn.CrossEntropyLoss()

loss=[]
n_epoch=[]

optimizer= torch.optim.SGD(net.parameters(), lr=0.0108)

for epoch in range(EPOCHS):
    y_pred= net(train_X)

    curr_loss= loss_fn(y_pred, train_y)

    optimizer.zero_grad()

    curr_loss.backward()

    optimizer.step()

    if epoch %100 == 0:
        loss.append(curr_loss.data.item())
            
        n_epoch.append(epoch)

    if epoch%1000 == 0:
        print("# of epochs: {0:05d} | Loss: {1:8.5f}".format(epoch, curr_loss))
```

```python
predict_out= net(test_X)

_, predict_y= torch.max(predict_out, 1)
```

```python
print ('prediction accuracy : {:.4f}'.format(accuracy_score(y_test.argmax(axis = 1), 
                                                            predict_y.data.cpu())))
```

```python
print(classification_report(y_test.argmax(axis = 1), predict_y.data.cpu()))
```

```python
loss_df=pd.DataFrame({'epoch': n_epoch,
                      'loss': loss})
```
