```python
def fn_render_env(name, act_size):
    
    env = gym.make(name, render_mode='rgb_array')
    #env.seed(RANDOM_STATE)
    state = env.reset(seed = RANDOM_STATE)
    
    img = plt.imshow(env.render())

    for j in range(STEPS):
        # action = agent.act(state)
        action = random.choice(range(act_size))
        img.set_data(env.render()) 
        #img.set_data(env.render()) 
        plt.axis('off')
        plt.title(name)
        display.display(plt.gcf())
        display.clear_output(wait=True)
        state, reward, done, truncated, info = env.step(action)
        #print (state, reward, done, truncated, info)
        if done:
            break 
        
    env.close()
    return 
```

## Understanding agent actions
The environments in gym have `Space` objects which describe the valid actions and observations. There fore two implimentations are `env.action_space` and `env.observation_space`

```python
env = gym.make('LunarLander-v2', render_mode='rgb_array')
observation_space = env.observation_space
action_space = env.action_space
reward_range = env.reward_range
print ('The Observation Space: {}'.format(observation_space))
print ('The Action Space: {}'.format(action_space))
print ('The Reward range Space: {}'.format(reward_range))
```

## Understand `env.step`
As mentioned in the documentation page, each environment is separated into different episodes, with `done=True` indicating that the specific episode has ended. Thus, we need to call reset there. For this, we need to understand what `env.step(action)` does and returns. `env.step(action)` takes the next step in the environment by performing the action specified by `action` and returns a tuple:
- done: 
- truncated : 
- info: Additional information for debugging


- observation (object)  this will be an element of the environments observation_space. This may, for instance, be a numpy array containing the positions and velocities of certain objects.This is environment specific and represents our observation of the environment after taking the action specified in `env.step(action)`.

- reward (float)  The amount of reward returned as a result of taking the action.

- terminated (bool)  whether a terminal state (as defined under the MDP of the task) is reached. In this case further step() calls could return undefined results. True if environment terminates (eg. due to task completion, failure etc.). We need to monitor this and call `env.reset()` when `done=True`.

- truncated (bool)  whether a truncation condition outside the scope of the MDP is satisfied. Typically a timelimit, but could also be used to indicate agent physically going out of bounds. Can be used to end the episode prematurely before a terminal state is reached. True if episode truncates due to a time limit or a reason that is not defined as part of the task MDP.

- info (dictionary)  info contains auxiliary diagnostic information (helpful for debugging, learning, and logging). This might, for instance, contain: metrics that describe the agents performance state, variables that are hidden from observations, or individual reward terms that are combined to produce the total reward. It also can contain information that distinguishes truncation and termination, however this is deprecated in favour of returning two booleans, and will be removed in a future version.
