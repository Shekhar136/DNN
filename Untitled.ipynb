```python
###Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.datasets import load_breast_cancer

###Global Parameters
#I/O Directory for storing Results
inpDir='../Inputs'
outDir='../Outputs'

#Setting the Random Seed
RANDOM_STATE=108
np.random.seed(RANDOM_STATE)

##Setting the Theme for all Graphs
sns.set_theme()

params={'legend.fontsize':18, 
        'font.family':'serif', 
        'font.size': 18, 
        'figure.figsize':(7,4), 
        'axes.labelsize': 20, 
        'axes.titlesize':22, 
        'xtick.labelsize':16,
        'ytick.labelsize':16
       }
plt.rcParams.update(params)
```

```python
#####################################################
##Saving Dataset to pd.DataFrame
dataset= load_breast_cancer()
print('-'*54)

##Setting X,y partitions
X=dataset.data

y=dataset.target

##Seeing the Dataset description
print(dataset.DESCR)
print('='*54)

###Splitting into Train and Test sets
X_train, X_test, y_train, y_test= train_test_split(X, y, 
                                                   test_size=0.2,
                                                   stratify=y,
                                                   random_state=RANDOM_STATE)
```

## Label [0,1]

```python
#####################################################
###Defining the Training Dataset
train_df=pd.DataFrame(X_train, columns= dataset.feature_names)
print('-'*54)
print(train_df.head(9))
print('-'*54)
##Giving an array and categorizing
pd.cut(np.array([1,5,2,3,6,4,6,8,7,7,7,2,4,3,5,9,9,4]), 2, labels=[0,1])

##Converting the Data to Binary format
X_bin_train= train_df.apply(pd.cut, bins=2, labels=[0,1])
print('-'*54)
print(X_bin_train.describe())
print('-'*54)
```

```python
#####################################################
##Plotting the Train Set with Subplots
fig, axes= plt.subplots(5,6)
#Axes Ravel gives us a LIST of graphs
axes=axes.ravel()
##Gives Matrix of 
for i, col in enumerate(X_bin_train.columns):
    X_bin_train[col].value_counts().plot(ax=axes[i], kind='bar', title=col)

plt.tight_layout()
```

```python
#####################################################
###Calculate a "Total" and add the column

X_bin_train['total']= X_bin_train.sum(axis=1)

##Adding a Target column
X_bin_train['target']= y_train

##Sorting X_bin_train
X_bin_train.sort_values('total', inplace=True)
print(f"The final X_Train looks as follows::\n {X_bin_train.head(9)}")
print('-'*54)
```

```python
#####################################################
##Plotting the X_bin to look at the values
tmp_df=X_bin_train.groupby(['total', 'target']).agg({'target':'count'})

tmp_df.columns=['count']
tmp_df.reset_index(inplace=True)

tmp_df.pivot(index='total',
             columns='target',
             values='count').plot(kind='bar')
plt.grid()
```

```python
#####################################################
##Accuracy of Model
accuracy={}

for b in range(X_bin_train.shape[1]):
    X_bin_train['y_pred']= X_bin_train['total']<=b
    
    accuracy[b]= accuracy_score(X_bin_train['y_pred'], X_bin_train['target'])

##Selecting the max parameter
best_b=min(accuracy, key=accuracy.get)
```

```python
best_b
```

```python
#####################################################
###Repeating the same for TEST SET
test_df=pd.DataFrame(X_test, columns= dataset.feature_names)
print('-'*54)
print(test_df.head(9))
print('-'*54)

##Converting the Data to Binary format
X_bin_test= test_df.apply(pd.cut, bins=2, labels=[0,1])
print('-'*54)
print(X_bin_test.describe())
print('-'*54)
```

```python
#####################################################
##Plotting the Test Set with Subplots
fig, axes= plt.subplots(5,6)
#Axes Ravel gives us a LIST of graphs
axes=axes.ravel()
 
for i, col in enumerate(X_bin_test.columns):
    X_bin_test[col].value_counts().plot(ax=axes[i], kind='bar', title=col)

plt.tight_layout()
```

```python
#####################################################
###Calculate a "Total" and add the column
X_bin_test['total']= X_bin_test.sum(axis=1)

##Adding a Target column
X_bin_test['target']= y_test

##Sorting X_bin_test
X_bin_test.sort_values('total', inplace=True)
print(f"The final X_Test looks as follows::\n {X_bin_test.head(9)}")
print('-'*54)
```

```python
#####################################################
##Plotting the X_bin to look at the values
tmp_df=X_bin_test.groupby(['total', 'target']).agg({'target':'count'})

tmp_df.columns=['count']
tmp_df.reset_index(inplace=True)

tmp_df.pivot(index='total',
             columns='target',
             values='count').plot(kind='bar')
plt.grid()
```

```python
#####################################################
X_bin_test['y_pred']=X_bin_test['total']<=best_b
accur= accuracy_score(X_bin_test['y_pred'], X_bin_test['target'])
```
